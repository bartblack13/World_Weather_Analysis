## Deliverable 3. Create a Travel Itinerary Map.

```python
# Dependencies and Setup
import pandas as pd
import requests
import gmaps

# Import API key
from config import g_key

# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# 1. Read the WeatherPy_vacation.csv into a DataFrame.
vacation_df = pd.read_csv("Vacation_Search/WeatherPy_vacation.csv")
vacation_df.head(10)
```

```python
# 2. Using the template add the city name, the country code, the weather description and maximum temperature for the city.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
<dt>Weather Description</dt><dd>{Weather Description}</dd>
</dl>
"""

# 3a. Get the data from each row and add it to the formatting template and store the data in a list.
hotel_info = [info_box_template.format(**row) for index, row in vacation_df.iterrows()]

# 3b. Get the latitude and longitude from each row and store in a new DataFrame.
locations = vacation_df[["Lat", "Lng"]]
```

```python
locations
```

```python
# 4a. Add a marker layer for each city to the map.
locations = vacation_df[["Lat", "Lng"]]

figure_layout = {
    'width': '800px',
    'height': '350px',
    'border': '1px solid black',
    'padding': '1px'}

fig = gmaps.figure(layout=figure_layout, center=(30.0, 31.0), zoom_level=1.5)

marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

fig.add_layer(marker_layer)

# 4b. Display the figure

fig
```

```python
# From the map above pick 4 cities and create a vacation itinerary route to travel between the four cities. 
# 5. Create DataFrames for each city by filtering the 'vacation_df' using the loc method. 
# Hint: The starting and ending city should be the same city.
# df.loc[df['column_name'] == some_value]
vacation_start = vacation_df.loc[vacation_df['Hotel Name'] == "Howard Johnson by Wyndham Plaza Hotel Mayorazgo"]
vacation_end = vacation_df.loc[vacation_df['Hotel Name'] == "Howard Johnson by Wyndham Plaza Hotel Mayorazgo"]
vacation_stop1 = vacation_df.loc[vacation_df['Hotel Name'] == "Hotel Provincial"]
vacation_stop2 = vacation_df.loc[vacation_df['Hotel Name'] == "Dallas Hotel Tucuman"] 
vacation_stop3 = vacation_df.loc[vacation_df['Hotel Name'] == "Apartamentos Castelli 90"] 
vacation_end
```

```python
# 6. Get the latitude-longitude pairs as tuples from each city DataFrame using the to_numpy function and list indexing.
# DataFrame.to_numpy(dtype=None, copy=False, na_value=NoDefault.no_default)
# subset = data_set[['data_date', 'data_1', 'data_2']]
# tuples = [tuple(x) for x in subset.to_numpy()]

start = [tuple(x) for x in vacation_start[['Lat', 'Lng']].to_numpy()]
end = [tuple(x) for x in vacation_end[['Lat', 'Lng']].to_numpy()]
stop1 = [tuple(x) for x in vacation_stop1[['Lat', 'Lng']].to_numpy()]
stop2 = [tuple(x) for x in vacation_stop2[['Lat', 'Lng']].to_numpy()]
stop3 = [tuple(x) for x in vacation_stop3[['Lat', 'Lng']].to_numpy()]
start[0]
```

```python
# 7. Create a direction layer map using the start and end latitude-longitude pairs,
# and stop1, stop2, and stop3 as the waypoints. The travel_mode should be "DRIVING", "BICYCLING", or "WALKING".
# import gmaps
# import gmaps.datasets
# gmaps.configure(api_key='AIza...')

# # Latitude-longitude pairs
# geneva = (46.2, 6.1)
# montreux = (46.4, 6.9)
# zurich = (47.4, 8.5)

# fig = gmaps.figure()
# geneva2zurich = gmaps.directions_layer(geneva, zurich)
# fig.add_layer(geneva2zurich)
# fig

import gmaps.datasets

figure_layout = {
    'width': '800px',
    'height': '350px',
    'border': '1px solid black',
    'padding': '1px'}

fig = gmaps.figure()
travel_route = gmaps.directions_layer(start[0], end[0], waypoints=[stop1[0], stop2[0], stop3[0]], travel_mode='DRIVING')
fig.add_layer(travel_route)
fig
```

```python
# 8. To create a marker layer map between the four cities.
#  Combine the four city DataFrames into one DataFrame using the concat() function.
destination_list = [vacation_start, 
                    vacation_stop1, 
                    vacation_stop2,
                    vacation_stop3]  
                    

itinerary_df = pd.concat(destination_list,ignore_index=True)
itinerary_df
```

```python
# 9 Using the template add city name, the country code, the weather description and maximum temperature for the city. 
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
<dt>Weather Description</dt><dd>{Weather Description}</dd>
</dl>
"""

# 10a Get the data from each row and add it to the formatting template and store the data in a list.
route_info = [info_box_template.format(**row) for index, row in itinerary_df.iterrows()]

# 10b. Get the latitude and longitude from each row and store in a new DataFrame.
locations2 = itinerary_df[["Lat", "Lng"]]
```

```python
# 11a. Add a marker layer for each city to the map.

# locations = vacation_df[["Lat", "Lng"]]; redundant, see above step 10b
figure_layout = {
    'width': '800px',
    'height': '350px',
    'border': '1px solid black',
    'padding': '1px'}

fig = gmaps.figure(layout=figure_layout, center=(-29.0, -64.0), zoom_level=5.8)

marker_layer = gmaps.marker_layer(locations2, info_box_content=route_info)

fig.add_layer(marker_layer)

# 11b. Display the figure
fig
```
